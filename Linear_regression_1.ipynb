```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('E:\\DATA SCIENCE\Competitions\Linear_regresssion\Linear_regresssion_exercise\canada_per_capita_income.csv')
df.head()
```

```python
df['per_capita_income'] = df['per capita income (US$)']
```

```python
df.drop('per capita income (US$)',axis=1 ,inplace=True)
```

```python
# Here we can see there is a linear relationship between the independent and dependent varaibles
plt.scatter(df[['year']], df[['per_capita_income']])
plt.xlabel('year')
plt.ylabel('per_capita_income')
plt.show()
```

```python
lr = LinearRegression()
```

```python
lr.fit(df[['year']], df.per_capita_income)
```

```python
year_pred ={'year' :[2016,2017,2018,2019,2020,2021,2022,2023,2024,2025,2026,2029]}
df_pred = pd.DataFrame(year_pred)
df_pred.head()
```

```python
predicted_results = lr.predict(df_pred)
df_pred['per_capita_income'] = pd.DataFrame(predicted_results)
df_pred.head()
```

```python
plt.scatter(df[['year']], df[['per_capita_income']])
plt.plot(df.year, lr.predict(df[['year']]))
plt.plot(df_pred.year, df_pred.per_capita_income, color = 'red')
plt.xlabel('year')
plt.ylabel('per_capita_income')
plt.show()
```

```python
# In y = mx + c, this is the m
m = lr.coef_
```

```python
# In y = mx + c, this is the c
c = (lr.intercept_)
```

```python
y = (m*2020) + c
print(y)
```
